```python
import re
punc = '[,.!\')(;:]'
import nltk
nltk.download('wordnet')
nltk.download('punkt')
from nltk.corpus import wordnet
from nltk.corpus import wordnet as wn
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = stopwords.words('english')
from bs4 import BeautifulSoup
```

```python
from gensim.models import KeyedVectors
word_vectors = KeyedVectors.load_word2vec_format('/home_lab_local/s1810410/WSD/GoogleNews-vectors-negative300.bin', binary=True)

re_synset = re.compile(r'[a-z]+\.[a-z]\.\d{2}')
noun_key_value = re.compile(r'[a-z]+[%][\d]+[:][\d]+[:][\d]+[:]+[a-z]*[:]*[\d\d]*')

f = open('EnglishLS.dictionary.mapping.xml', 'r')
soup= BeautifulSoup(f) 
```

```python
class Get_gloss:
  # 接受一个sense-key， 返回他的词义的上位词， 下位词， 近义词的词典信息 
    def get_hyp_sense(self,key_value):
    
        lema = str(wn.lemma_from_key(key_value))
        r_lema = re.search(re_synset,lema).group()
        syn_r_lema = wn.synset(r_lema)
        hypon_syn = syn_r_lema.hyponyms()
        hyper_syn = syn_r_lema.hypernyms()
        member_syn = syn_r_lema.member_holonyms()
        
        full_text = [] 
        if len(hypon_syn) != 0:
            one =  str(hypon_syn[0])
            r_one = re.search(re_synset,one).group()
            sense_one = wn.synset(r_one).definition()
            text_list = nltk.word_tokenize(sense_one)
            text_list = [word for word in text_list if word in word_vectors.wv.vocab]
            text_list = [word for word in text_list if word not in stop_words]
            text_list = [word for word in text_list if word not in punc]
            for word in text_list:
                full_text.append(word)
                
        if len(hyper_syn) != 0:
            two =  str(hyper_syn[0])
            r_two = re.search(re_synset,two).group()
            sense_two = wn.synset(r_two).definition()
            text_list2 = nltk.word_tokenize(sense_two)
            text_list2 = [word for word in text_list2 if word in word_vectors.wv.vocab]
            text_list2 = [word for word in text_list2 if word not in stop_words]
            text_list2 = [word for word in text_list2 if word not in punc]
            for word in text_list2:
                full_text.append(word)

        if len(member_syn) != 0:
            three = str(member_syn[0])
            r_three = re.search(re_synset,three).group()
            sense_three = wn.synset(r_three).definition()
            text_list3 = nltk.word_tokenize(sense_three)
            text_list3 = [word for word in text_list3 if word in word_vectors.wv.vocab]
            text_list3 = [word for word in text_list3 if word not in stop_words]
            text_list3 = [word for word in text_list3 if word not in punc]
            for word in text_list3:
                full_text.append(word)
        return full_text

    def get_key_sense(self,key_value):
        ## ( get gloss from key value )
        lemma = str(wn.lemma_from_key(key_value))
        r_lema = re.search(re_synset,lemma).group()
        sense = wn.synset(r_lema).definition()
        text_list = nltk.word_tokenize(sense)
        text_list = [word for word in text_list if word in word_vectors.wv.vocab]
        text_list = [word for word in text_list if word not in stop_words]
        text_list = [word for word in text_list if word not in punc]
        return text_list

GL = Get_gloss()
```
